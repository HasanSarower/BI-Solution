```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
from datetime import datetime
%matplotlib inline
```

```python
#read CSV file
df_raw = pd.read_csv("raw.csv")
```

```python
df_raw = df_raw.fillna(0)
```

```python
df_raw = df_raw.set_index(df_raw.Timestamp)
```

```python
df_raw['Year'] = df_raw.index.year
df_raw['Month'] = df_raw.index.month
df_raw['day'] = df_raw.index.day
df_raw['Time'] = df_raw.index.time
df_raw.head()
```

```python
import matplotlib.dates as mdates
```

```python
data_columns = ['Open','Close', 'Volume_(BTC)', 'Volume_(Currency)','Weighted_Price']
data_raw_weekly = df_raw[data_columns].resample('W').mean()
data_raw_daily = df_raw[data_columns].resample('D').mean()
data_raw_Monthly = df_raw[data_columns].resample('M').mean()
```

```python
# Start and end of the date range to extract
start, end = '2018-01', '2019-08'
# Plot daily and weekly resampled time series together
fig, ax = plt.subplots()
ax.plot(data_raw_daily.loc[start:end, 'Close'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(data_raw_weekly.loc[start:end, 'Close'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Close')
ax.legend();
```

```python
#Now explore the monthly time series
fig, ax = plt.subplots()
ax.plot(data_raw_Monthly['Close'], color='black', label='Close price')
data_raw_Monthly[['Open', 'Volume_(Currency)','Volume_(BTC)']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Close');
```

```python
#Rolling window
df_1d = df_raw[data_columns].rolling(1, center=True).mean()  
```

```python
# Start and end of the date range to extract
start, end = '2014-01', '2019-08'
# Plot daily, weekly resampled, and 1-day rolling mean time series together
fig, ax = plt.subplots()
ax.plot(df_raw.loc[start:end, 'Close'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(data_raw_weekly.loc[start:end, 'Close'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.plot(df_1d.loc[start:end, 'Close'],
marker='.', linestyle='-', label='1-d Rolling Mean')
ax.set_ylabel('Close price)')
ax.legend();
```

```python
#trends
df_365d = df_raw[data_columns].rolling(window=365, center=True, min_periods=360).mean()
df_3d = df_raw[data_columns].rolling(3, center=True).mean()
```

```python
# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plt.subplots()
ax.plot(data_raw_daily['Close'], marker='.', markersize=2, color='0.6',
linestyle='None', label='Daily')
ax.plot(df_3d['Close'], linewidth=2, label='3-d Rolling Mean')
ax.plot(df_365d['Close'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Close price')
ax.set_title('Trends in Close price');
```

```python
#trend in open,close and Weighted_Price
fig, ax = plt.subplots()
for nm in ['Open', 'Close', 'Weighted_Price']:
    ax.plot(df_365d[nm], label=nm)
    # Set x-ticks to yearly interval, add legend and labels
    ax.xaxis.set_major_locator(mdates.YearLocator())
    
    ax.legend()
    ax.set_ylabel('Close)')
    ax.set_title('')
```
